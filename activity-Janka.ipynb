```python
norm = stats.norm(0, 1)
x = data.Open
sample = norm.rvs(data.Open.size)

plt.plot(x, norm.pdf(x))
plt.hist(sample, bins=20)
plt.title("Normal distribution of Open: \nSkewness %.5f, Kurtosis %.5f" % (stats.skew(sample), stats.kurtosis(sample, fisher=False)))
```

```python
chi2 = stats.chi2(5)
x = data.Open
sample = chi2.rvs(data.Open.size)

plt.plot(x, chi2.pdf(x))
plt.hist(sample, bins=20)
plt.title("Chi-squared distribution of Open : ""Skewness %.5f" % (stats.skew(sample)))
```

```python
logistic = stats.logistic()
x = data.Open
sample = logistic.rvs(data.Open.size)

plt.plot(x, logistic.pdf(x))
plt.hist(sample, bins=20)

plt.title("Logistic distribution Open: ""Kurtosis %.5f" % (stats.kurtosis(sample, fisher=False)))
```

```python
uniform = stats.uniform()
x = data.Open
sample = uniform.rvs(data.Open.size)

plt.plot(x, uniform.pdf(x))
plt.hist(sample, bins=20)

plt.title("Uniform distribution Open: ""Kurtosis %.5f" % (stats.kurtosis(sample, fisher=False)))
```
